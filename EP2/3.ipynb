```python
class Account:
    def __init__(self, name, amount):
        self.name = name
        self.amount = amount

    def withdraw(self, amount):
        self.amount -= amount

    def deposit(self, amount):
        self.amount += amount

class SavingsAccount(Account):
    def __init__(self, name, amount, interest_rate):
        super().__init__(name, amount)
        self.interest_rate = interest_rate

    def interest(self):
        self.amount += self.amount * self.interest_rate

class PremiumSavingsAccount(SavingsAccount):
    def __init__(self, name, amount, interest_rate, insurance_amount, atm_limit, credit_card_balance):
        super().__init__(name, amount, interest_rate)
        self.insurance_amount = insurance_amount
        self.atm_limit = atm_limit
        self.credit_card_balance = credit_card_balance

class LoanAccount(Account):
    def __init__(self, name, amount, interest_rate, loan_term, fees_and_charges):
        super().__init__(name, amount)
        self.interest_rate = interest_rate
        self.loan_term = loan_term
        self.fees_and_charges = fees_and_charges

class CurrentAccount(Account):
    def __init__(self, name, amount, short_term_funds_minimum_balance, penalty):
        super().__init__(name, amount)
        self.short_term_funds_minimum_balance = short_term_funds_minimum_balance
        self.penalty = penalty

    def penalty_calculation(self):
        if self.amount < self.short_term_funds_minimum_balance:
            self.penalty = (self.short_term_funds_minimum_balance - self.amount) * 0.02
```
