```python
import datetime

class Book:
   def __init__(self, title, author, categories):
       self.title = title
       self.author = author
       self.categories = categories
       self.availability = "Yes"

   def set_availability(self, availability):
       self.availability = availability

   def get_availability(self):
       return self.availability

class Member:
   def __init__(self, name, contact_detail):
       self.name = name
       self.contact_detail = contact_detail

class Loan:
   def __init__(self, book, member, borrowed_date, return_date):
       self.book = book
       self.member = member
       self.borrowed_date = borrowed_date
       self.return_date = return_date

class Library:
   def __init__(self):
       self.books = []
       self.members = []
       self.loans = []

   def add_book(self, book):
       self.books.append(book)

   def add_member(self, member):
       self.members.append(member)

   def issue_book(self, book_title, member_name, borrowed_date, return_date):
       book = None
       member = None

       for b in self.books:
           if b.title == book_title:
               book = b
               break

       for m in self.members:
           if m.name == member_name:
               member = m
               break

       if book is None or member is None:
           return "Book or member not found"

       if book.get_availability() == "No":
           return "Book is not available"

       loan = Loan(book, member, borrowed_date, return_date)
       self.loans.append(loan)
       book.set_availability("No")

       return "Book issued successfully"

   def return_book(self, book_title, member_name):
       book = None
       member = None

       for l in self.loans:
           if l.book.title == book_title and l.member.name == member_name:
               book = l.book
               member = l.member
               break

       if book is None or member is None:
           return "Book or member not found"

       book.set_availability("Yes")
       self.loans.remove(l)

       return "Book returned successfully"

library = Library()

book1 = Book("Python for Beginners", "John Smith", ["Programming", "Python"])
book2 = Book("Learn Java in 30 Days", "Mike Johnson", ["Programming", "Java"])
book3 = Book("The C Programming Language", "David Malan", ["Programming", "C"])

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

member1 = Member("Alice Johnson", "1234567890")
member2 = Member("Bob Davis", "2345678901")

library.add_member(member1)
library.add_member(member2)

print(library.issue_book("Python for Beginners", "Alice Johnson", datetime.date(2022, 1, 1), datetime.date(2022, 1, 14)))
print(library.issue_book("Learn Java in 30 Days", "Bob Davis", datetime.date(2022, 1, 1), datetime.date(2022, 1, 14)))
print(library.issue_book("The C Programming Language", "Alice Johnson", datetime.date(2022, 1, 1), datetime.date(2022, 1, 14)))
print(library.return_book("Python for Beginners", "Alice Johnson"))
print(library.return_book("The C Programming Language", "Alice Johnson"))
```
