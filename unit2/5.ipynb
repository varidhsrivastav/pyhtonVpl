```python
class BaseCharacter:
    def __init__(self, name):
        self._name = name
        self._weapon = None
        BaseCharacter._count += 1

    @property
    def name(self):
        return self._name

    @property
    def weapon(self):
        return self._weapon

    @weapon.setter
    def weapon(self, new_weapon):
        self._weapon = new_weapon

    @classmethod
    def display_count(cls):
        print(f"Number of characters created: {cls._count}")

    @staticmethod
    def game_rules():
        print("Welcome to the game! Follow the rules and enjoy the adventure.")

    def attack(self):
        raise NotImplementedError("Subclasses must implement the attack method.")


class Wizard(BaseCharacter):
    _count = 0

    def __init__(self, name):
        super().__init__(name)
        self._magic_power = 50

    @property
    def magic_power(self):
        return self._magic_power

    def attack(self):
        print(f"{self._name} casts a spell using {self._weapon} with magic power {self._magic_power}.")


class Warrior(BaseCharacter):
    _count = 0

    def __init__(self, name):
        super().__init__(name)
        self._strength = 70

    @property
    def strength(self):
        return self._strength

    def attack(self):
        print(f"{self._name} attacks with {self._weapon} using strength {self._strength}.")


class Archer(BaseCharacter):
    _count = 0

    def __init__(self, name):
        super().__init__(name)
        self._precision = 80

    @property
    def precision(self):
        return self._precision

    def attack(self):
        print(f"{self._name} shoots arrows with {self._weapon} using precision {self._precision}.")


class Runner(BaseCharacter):
    _count = 0

    def __init__(self, name):
        super().__init__(name)
        self._speed = 100

    @property
    def speed(self):
        return self._speed

    def attack(self):
        print(f"{self._name} runs swiftly with {self._weapon} at a speed of {self._speed}.")


if __name__ == "__main__":
    wizard1 = Wizard("Gandalf")
    wizard1.weapon = "Staff"
    wizard1.attack()

    warrior1 = Warrior("Aragorn")
    warrior1.weapon = "Sword"
    warrior1.attack()

    archer1 = Archer("Legolas")
    archer1.weapon = "Bow"
    archer1.attack()

    runner1 = Runner("Usain Bolt")
    runner1.weapon = "Sneakers"
    runner1.attack()

    BaseCharacter.display_count()
    BaseCharacter.game_rules()
```
