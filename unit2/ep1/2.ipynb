```python
class BaseCharacter:
    _count = 0
    def __init__(self, name):
        self._name = name
        self._weapon = None
        BaseCharacter._count += 1

    @property
    def name(self):
        return self._name

    @property
    def weapon(self):
        return self._weapon

    @weapon.setter
    def weapon(self, new_weapon):
        self._weapon = new_weapon

    @classmethod
    def display_count(cls):
        print(f"Number of characters created: {cls._count}")

    @staticmethod
    def game_rules():
        print("Welcome to the game!")

    def attack(self):
        pass


class Wizard(BaseCharacter):
    _count = 0
    def __init__(self, name):
        super().__init__(name)
        self._magic_power = 50

    @property
    def magic_power(self):
        return self._magic_power

    def attack(self):
        print(f"{self._name} create portals {self._weapon} -- magic power {self._magic_power}.")


class Warrior(BaseCharacter):
    _count = 0

    def __init__(self, name):
        super().__init__(name)
        self._strength = 70

    @property
    def strength(self):
        return self._strength

    def attack(self):
        print(f"{self._name} attaked with iron suites {self._weapon} using strength {self._strength}.")

class Archer(BaseCharacter):
    _count = 0

    def __init__(self, name):
        super().__init__(name)
        self._precision = 80

    @property
    def precision(self):
        return self._precision

    def attack(self):
        print(f"{self._name} attacked with happer {self._weapon} using precision {self._precision}.")


class Runner(BaseCharacter):
    _count = 0
    def __init__(self, name):
        super().__init__(name)
        self._speed = 100
    @property
    def speed(self):
        return self._speed

    def attack(self):
        print(f"{self._name} runs swiftly  {self._weapon} at a speed of {self._speed}.")


wizard1 = Wizard("doctor strange")
wizard1.weapon = "time stone"
wizard1.attack()

warrior1 = Warrior("iron man ")
warrior1.weapon = "armor suit"
warrior1.attack()

archer1 = Archer("thor")
archer1.weapon = "hammer"
archer1.attack()

runner1 = Runner("natasha romanaoff")
runner1.weapon = "electricity"
runner1.attack()

BaseCharacter.display_count()
BaseCharacter.game_rules()
```

```python
class Animal:
    _count = 0
    def __init__(self, name):
        self._name = name
        Animal._count += 1
    @property
    def name(self):
        return self._name
    @classmethod
    def display_count(cls):
        print(f"instances: {cls._count}")
    def sound(self):
        pass


class Mammal(Animal):
    _count = 0
    def __init__(self, name):
        super().__init__(name)
        Mammal._count += 1
    def give_birth(self):
        print(f"{self._name} give birth")

    def sound(self):
        print(f"{self._name} makes m sounds.")


class Bird(Animal):
    _count = 0
    def __init__(self, name):
        super().__init__(name)
        Bird._count += 1
    def give_birth(self):
        print(f"{self._name} gave birth.")


class Bat(Mammal):
    _count = 0
    def __init__(self, name):
        super().__init__(name)
        Bat._count += 1
    def move(self):
        print(f"{self._name} moves.")

class Ostrich(Bird):
    _count = 0
    def __init__(self, name):
        super().__init__(name)
        Ostrich._count += 1
    def move(self):
        print(f"{self._name} move.")



mammal1 = Mammal("Dog")
mammal1.sound()
mammal1.give_birth()

bird1 = Bird("Eagle")
bird1.sound()
bird1.give_birth()

bat1 = Bat("Bat")
bat1.sound()
bat1.move()

ostrich1 = Ostrich("ostrich")
ostrich1.sound()
ostrich1.move()

Animal.display_count()
Mammal.display_count()
Bird.display_count()
Bat.display_count()
Ostrich.display_count()
```

```python
class Person:
    total_created = 0
    adults = 0
    minors = 0

    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.total_created += 1

        if self.is_adult():
            Person.adults += 1
        else:
            Person.minors += 1

    @classmethod
    def deduce_age(cls, name, birth_year, present_year):
        age = present_year - birth_year
        person = cls(name, age)
        print(f"The deduced age of {person.name} is {person.age}.")

    def is_adult(self):
        return self.age >= 18

# Example usage
present_year = 2024

p1 = Person("ironman", 35)
p2 = Person("thor", 25)
p3 = Person("hulk", 15)


print(f"{p1.name} {'is not ' if not p1.is_adult() else 'is'} an adult")
print(f"{p2.name} {'is not ' if not p2.is_adult() else 'is'} an adult")
print(f"{p3.name} {'is not ' if not p3.is_adult() else 'is'} an adult")

print("\nTotal persons: ", Person.total_created)
print("Total adults: ", Person.adults)
print("Total minors: ", Person.minors)
```

```python
class BaseClass:
    def __init__(self, name):
        self.name = name
        self.weapon_of_char = None

    def arms(self):
        return self.weapon_of_char

    def set_arms(self, arms):
        self.weapon_of_char = arms


class Wizard(BaseClass):
    def __init__(self, name):
        super()._init_(name)
        self.weapon_of_char = "Magic"

    def attack(self):
        return f"{self.name} Casting magical spell"


class Warrior(BaseClass):
    def __init__(self, name):
        super()._init_(name)
        self.weapon_of_char = "Fighter"

    def attack(self):
        return f"{self.name} punches!!"


class Archer(BaseClass):
    def __init__(self, name):
        super()._init_(name)
        self.weapon_of_char = "Bow N Arrow"

    def attack(self):
        return f"{self.name} aims at the target and releases an arrow"


class Runner(BaseClass):
    def __init__(self, name):
        super()._init_(name)
        self.weapon_of_char = "Javelin Dagger"

    def attack(self):
        return f"{self.name} aims and throws"


class GameChar(BaseClass):
    def __init__(self, name, char_type):
        super()._init_(name)
        self.char_type = char_type

    def intro(self):
        return f"I am a {self.char_type} known as {self.name}."

    def exp_char_arms(self):
        if self.char_type.lower() == "wizard":
            return Wizard.attack(self)
        elif self.char_type.lower() == "warrior":
            return Warrior.attack(self)
        elif self.char_type.lower() == "archer":
            return Archer.attack(self)
        elif self.char_type.lower() == "runner":
            return Runner.attack(self)


# Example Usage:

wizard = GameChar("PROF SNAPE", "wizard")
warrior = GameChar("HANUMAN ji", "warrior")
archer = GameChar("SHREE RAM", "archer")
runner = GameChar("flash", "runner")

print(wizard.exp_char_arms())
print(warrior.exp_char_arms())
print(archer.exp_char_arms())
print(runner.exp_char_arms())
```
