#static method


```python
class myclass:
    v=10
    def __init__(self,val):
        self.val = val
    @staticmethod
    def f(a,b):
        return a+b
myclass.f(10,20)
type(myclass.f)
```

```python

class maths:
    @staticmethod
    def is_even(x):
        return x % 2 == 0
    @staticmethod
    def pow(x,y):
        return(x**y)
    @staticmethod
    def fact(a):
        if  a<=0:
            return 1
        else:
            return (a*maths.fact(a-1))
    @staticmethod
    def sub(x,y):
        return(x-y)
maths.fact(10)   
maths.is_even(23) 
print("Power of 5 to 4 : ",maths.pow(5,4))

m=maths()
m.sub(50,20)
```

```python
class myclass:
    v=10
    @classmethod
    def get_value(cls,x):
        cls.v+=x
myclass.get_value(4)
myclass.v
```

```python
class animal:
    def __init__(self, breed):
        self.breed = breed
    @classmethod
    def create_from_breed(cls, breed):
        aa=cls(breed)
        return cls(aa)
a=animal("dog")
a.breed
type(a)
```
